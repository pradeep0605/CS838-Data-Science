```python
import py_entitymatching as em
import pandas as pd
import os

# Read the CSV files
goodreads_path = "/Users/sowrabhahg/Coursework/cs838/Project/CS838-Data-Science/Phase3/test/goodreads.csv"
barnes_path = "/Users/sowrabhahg/Coursework/cs838/Project/CS838-Data-Science/Phase3/test/barnes_and_nobles_crawl.csv"
goodreads= em.read_csv_metadata(goodreads_path, key='ISBN13')
barnes = em.read_csv_metadata(barnes_path, key='ISBN13')
```

```python
# Instantiate Attr. Equivalence Blocker
ab = em.AttrEquivalenceBlocker()
# Use block_tables to apply blocking over two input tables.
C2 = ab.block_tables(goodreads, barnes, 
                    l_block_attr='ISBN13', r_block_attr='ISBN13', 
                    l_output_attrs=['Original_Title', 'Author', 'ISBN13', 'Publisher', 'Published_Date'],
                    r_output_attrs=['Original_Title', 'Author', 'ISBN13', 'Publisher', 'Published_Date'])
```

```python
# Specify the tokenization to be 'word' level and set overlap_size to be 3.
# Instantiate overlap blocker object
ob = em.OverlapBlocker()
C1 = ob.block_tables(goodreads, barnes, 'Original_Title', 'Original_Title', word_level=True, overlap_size=5, 
                    l_output_attrs=['Original_Title', 'Author', 'ISBN13', 'Publisher', 'Published_Date'], 
                    r_output_attrs=['Original_Title', 'Author', 'ISBN13', 'Publisher', 'Published_Date'])
C1
```

```python
C3 = ab.block_tables(goodreads, barnes, 
                    l_block_attr='Original_Title', r_block_attr='Original_Title', 
                    l_output_attrs=['Original_Title', 'Author', 'ISBN13', 'Publisher', 'Published_Date'],
                    r_output_attrs=['Original_Title', 'Author', 'ISBN13', 'Publisher', 'Published_Date'])
C3
```

```python
C4 = ab.block_tables(goodreads, barnes, 
                    l_block_attr='Author', r_block_attr='Author', 
                    l_output_attrs=['Original_Title', 'Author', 'ISBN13', 'Publisher', 'Published_Date'],
                    r_output_attrs=['Original_Title', 'Author', 'ISBN13', 'Publisher', 'Published_Date'])
```

```python
C = em.combine_blocker_outputs_via_union([C1,C2,C3,C4])
em.debug_blocker(candset=C,ltable=goodreads,rtable=barnes)
```

```python
block_f = em.get_features_for_blocking(goodreads, barnes)
rb = em.RuleBasedBlocker()
# Add rule : block tuples if name_name_lev(ltuple, rtuple) < 0.4
rb.add_rule(['Author_Author_lev_sim(ltuple, rtuple) < 0.8'], block_f)
G = rb.block_candset(C, show_progress=False)
```

```python
G1 = ob.block_candset(G, 'Original_Title', 'Original_Title', word_level=True, overlap_size=1, show_progress=False, rem_stop_words=True)
G1
```

```python
Labled_Data = em.read_csv_metadata('S.csv', key='_id',ltable=goodreads, rtable=barnes,fk_ltable='ltable_ISBN13', fk_rtable='rtable_ISBN13')
```

```python
# Split S into I an J
IJ = em.split_train_test(Labled_Data, train_proportion=0.5, random_state=0)
I = IJ['train']
J = IJ['test']
```

```python
# Create a set of ML-matchers
dt = em.DTMatcher(name='DecisionTree', random_state=0)
svm = em.SVMMatcher(name='SVM', random_state=0)
rf = em.RFMatcher(name='RF', random_state=0)
lg = em.LogRegMatcher(name='LogReg', random_state=0)
ln = em.LinRegMatcher(name='LinReg')
```

```python
# Generate a set of features
#goodreads.head()
#barnes.head()
F = em.get_features_for_matching(goodreads, barnes)
```

```python
sim = em.get_sim_funs_for_matching()
tok = em.get_tokenizers_for_matching()
#feature_string = """jaccard((ltuple['Original_Title']).lower(), (rtuple['Original_Title']).lower())"""
feature_string = """jaccard(wspace((ltuple['Original_Title'].lower())), 
                            wspace((rtuple['Original_Title'].lower())))"""
feature = em.get_feature_fn(feature_string, sim, tok)

# Add feature to F
em.add_feature(F, 'jac_ws_name_address', feature)
```

```python
# Convert the I into a set of feature vectors using F
H = em.extract_feature_vecs(I, 
                            feature_table=F, 
                            attrs_after='label',
                            show_progress=False)  
```

```python
# Select the best ML matcher using CV
result = em.select_matcher([dt, rf, svm, ln, lg], table=H, 
        exclude_attrs=['_id', 'ltable_ISBN13', 'rtable_ISBN13', 'label'],
        k=5,
        target_attr='label', metric='f1', random_state=0)
result['cv_stats']
```

```python
# Train using feature vectors from I 
ln.fit(table=H, 
       exclude_attrs=['_id', 'ltable_ISBN13', 'rtable_ISBN13', 'label'], 
       target_attr='label')
```

```python
# Convert J into a set of feature vectors using F
L = em.extract_feature_vecs(J, feature_table=F,
                            attrs_after='label', show_progress=False)
```

```python
L
```

```python
# Predict on L 
predictions = ln.predict(table=L, exclude_attrs=['_id', 'ltable_ISBN13', 'rtable_ISBN13', 'label'], 
              append=True, target_attr='predicted', inplace=False)
```

```python
# Evaluate the predictions
eval_result = em.eval_matches(predictions, 'label', 'predicted')
em.print_eval_summary(eval_result)
```
