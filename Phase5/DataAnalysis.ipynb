```python
import pandas as pd
import numpy as np

df = pd.read_csv('./merged_table.csv')
df
```

```python
#Null Values of Genres
df[pd.isnull(df['Genres'])]
genre_null_count = len(df[pd.isnull(df['Genres'])])
print genre_null_count
```

```python
#remove rows with null genre
df_genre_filtered = df[pd.notnull(df['Genres'])]
df_genre_filtered
```

```python
#Convert Genres column from string to list of strings
#splitter = lambda x: x.split(', ')
split_n_strip = lambda x : [y.strip() for y in x.split(',')]
df_genre_filtered['Genres'] = df_genre_filtered['Genres'].apply(split_n_strip)

df_genre_filtered
```

```python
s = df_genre_filtered.head(n = 3)

#Try to get a unique set of groups for Genres
s_genres_list= s['Genres'].tolist()
print s_genres_list
s_genres = set()
for s_g_list in s_genres_list :
    for s_genre in s_g_list:
        s_genres.add(s_genre)

print s_genres
print len(s_genres)
```

```python
# Get the unique set of genres for the whole table
genres_list= df_genre_filtered['Genres'].tolist()

# Get only unique instance of a genre in a list
gen_lst_rem_rep = []
for gen_list in genres_list:
    gen_lst_rem_rep.append(set(gen_list))
    
#print genres_list
genres = set()
for g_list in genres_list :
    for genre in g_list:
        genres.add(genre.strip())

print genres
print len(genres)
```

```python
#Number of books in each genre
genre_dict = {}
for g_list in gen_lst_rem_rep :
    for genre in g_list:
        if genre in genre_dict:
            genre_dict[genre] +=1
        else:
            genre_dict[genre] = 1
            
#sort by values descending
rev_sorted_g = sorted(genre_dict, key=genre_dict.get, reverse=True)

#print the results of the dictionary sorted in reverse order
for g in rev_sorted_g:
    print g + "        :" +  str(genre_dict[g])
```

```python
#pick top 10 genres
top_10_g = rev_sorted_g[:10]

#print the top 10 genres
for g in top_10_g:
    print g + "        :" +  str(genre_dict[g])
```

```python
#sort by values ascending
sorted_g = sorted(genre_dict, key=genre_dict.get)

#pick top 10 genres
bottom_10_g = sorted_g[:10]

#print the top 10 genres
for g in bottom_10_g:
    print g + "        :" +  str(genre_dict[g])
```

```python
#Perform analysis on top 5 genres
top_5_g = top_10_g[:5]
#print the top 5 genres
for g in top_5_g:
    print g + "        :" +  str(genre_dict[g])
```

```python
#Create a table for each genre in top 5 genres
#Fiction Dataframe
fiction_filter = lambda x: top_5_g[0] in x
df_fiction = df_genre_filtered.loc[df_genre_filtered['Genres'].apply(fiction_filter)]

#Drop Genres column because it is a list
df_fiction = df_fiction.drop('Genres', 1)

#Add Genre colun with only 'Fiction' for fiction dataframe
df_fiction['Genre']= top_5_g[0]

#Just take the first word of publisher
strp = lambda x: x.strip()
pub_sub = lambda x: x[:x.find(' ')]
#Clean if the publisher name contains harper collins
clean_harp = lambda x: 'HarperCollins' if x.find('Harper') != -1 else x
df_fiction['Publication'] = df_fiction['Publication'].apply(strp)
df_fiction['Publication'] = df_fiction['Publication'].apply(pub_sub)
df_fiction['Publication'] = df_fiction['Publication'].apply(clean_harp)

print len(df_fiction)
df_fiction.head()
```

```python
#Fantasy Dataframe
fantasy_filter = lambda x: top_5_g[1] in x
df_fantasy = df_genre_filtered.loc[df_genre_filtered['Genres'].apply(fantasy_filter)]

#Drop Genres column because it is a list
df_fantasy = df_fantasy.drop('Genres', 1)

df_fantasy['Publication'] = df_fantasy['Publication'].apply(strp)
df_fantasy['Publication'] = df_fantasy['Publication'].apply(pub_sub)
df_fantasy['Publication'] = df_fantasy['Publication'].apply(clean_harp)


#Add Genre colun with only 'Fantasy' for Fantasy dataframe
df_fantasy['Genre']= top_5_g[1]
print len(df_fantasy)
df_fantasy.head()
```

```python
#Young Adult Dataframe
y_adult_filter = lambda x: top_5_g[2] in x
df_y_adult = df_genre_filtered.loc[df_genre_filtered['Genres'].apply(y_adult_filter)]

#Drop Genres column because it is a list
df_y_adult = df_y_adult.drop('Genres', 1)

df_y_adult['Publication'] = df_y_adult['Publication'].apply(strp)
df_y_adult['Publication'] = df_y_adult['Publication'].apply(pub_sub)
df_y_adult['Publication'] = df_y_adult['Publication'].apply(clean_harp)

#Add Genre colun with only 'Young Adult' for Young Adult dataframe
df_y_adult['Genre']= top_5_g[2]
print len(df_y_adult)
df_y_adult.head()
```

```python
#Romance Dataframe
romance_filter = lambda x: top_5_g[3] in x
df_romance = df_genre_filtered.loc[df_genre_filtered['Genres'].apply(romance_filter)]

#Drop Genres column because it is a list
df_romance = df_romance.drop('Genres', 1)

df_romance['Publication'] = df_romance['Publication'].apply(strp)
df_romance['Publication'] = df_romance['Publication'].apply(pub_sub)
df_romance['Publication'] = df_romance['Publication'].apply(clean_harp)

#Add Genre colun with only 'Romance' for Romance dataframe
df_romance['Genre']= top_5_g[3]
print len(df_romance)
df_romance.head()
```

```python
#Contemporary Dataframe
cont_filter = lambda x: top_5_g[4] in x
df_cont = df_genre_filtered.loc[df_genre_filtered['Genres'].apply(cont_filter)]

#Drop Genres column because it is a list
df_cont = df_cont.drop('Genres', 1)

df_cont['Publication'] = df_cont['Publication'].apply(strp)
df_cont['Publication'] = df_cont['Publication'].apply(pub_sub)
df_cont['Publication'] = df_cont['Publication'].apply(clean_harp)

#Add Genre colun with only 'Contemporary' for Contemporary dataframe
df_cont['Genre']= top_5_g[4]
print len(df_cont)
df_cont.head()
```

```python
#Calculate the average rating of each genre
avg_rtg = {}
print "Genre : Average Rating"

genre_df_dict = {top_5_g[0]:df_fiction, top_5_g[1]:df_fantasy, top_5_g[2]:df_y_adult, top_5_g[3]:df_romance, top_5_g[4]:df_cont}

for dfg in top_5_g:
    #print dfg['Average_Rating'].mean()
    avg_rtg[dfg] = genre_df_dict[dfg]['Average_Rating'].mean()

for g in top_5_g:
    print g + " : " + str(avg_rtg[g])
```

```python
#pub_count = df_fiction.groupby(['Publication']).size()
#print pub_count

#pub_count.to_csv('./fiction_grp_cnt.csv')

pub_group_count = {}
for dfg in top_5_g:
    pub_group_count[dfg] = genre_df_dict[dfg].groupby(['Publication']).size()
    
for g in top_5_g:
    pub_group_count[g].to_csv('./PublisherCount/' + g + '_pub_grp_cnt.csv')
```

```python
#Years in which most of the books in each genre were published
year_group_count = {}
for dfg in top_5_g:
    year_group_count[dfg] = genre_df_dict[dfg].groupby(['Published_Date']).size()
    
for g in top_5_g:
    year_group_count[g].to_csv('./YearCount/' + g + '_y_grp_cnt.csv')
```

```python
#Overall Number of Ratings per year
#Convert ratings from string to int
rtg_to_int = lambda x: int(x.replace(',', ''))
df_genre_filtered['Ratings'] = df_genre_filtered['Ratings'].apply(rtg_to_int)

groupby_object = df_genre_filtered.groupby(['Published_Date'])

year_stats = groupby_object.agg('sum').rename(columns = lambda x: x + ' sum').join(pd.DataFrame(groupby_object.size(), columns=['counts']))

year_stats.to_csv('yearly_stats.csv')
```

```python
lang_groupby_object = df_genre_filtered.groupby(['Language'])

lang_stats = lang_groupby_object.agg('mean').rename(columns = lambda x: x + ' mean').join(pd.DataFrame(lang_groupby_object.size(), columns=['counts']))

lang_stats.to_csv('stats_by_lang.csv')
```
